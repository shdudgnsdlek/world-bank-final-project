# Beyond GDP: Exploring Balanced and Unbalanced Development Across Countries

## 1. Introduction

### This project explores whether national development can be explained by GDP alone. It analyzes economic, social, technological, infrastructural, labor, and environmental indicators. The goal is to understand whether countries develop in a balanced or unbalanced way.

## 2. Dataset Description

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv(r"C:\Users\nyh11\Downloads\P_Data_Extract_From_World_Development_Indicators\world_bank_development_indicators.csv")
```

```python
df.columns = df.columns.str.strip()

print(df.columns)
display(df.head(10))
```

## 3. Data Cleaning and Preparation

```python
df.columns = df.columns.str.replace(r"\s*\[YR\d{4}\]", "", regex=True)

print(df.columns)
display(df.head())
```

```python
id_cols = ["Country Name", "Country Code", "Series Name", "Series Code"]

year_cols = [col for col in df.columns if col.isdigit()]

print(year_cols[:5])
print(year_cols[-5:])
```

```python
world_long = df.melt(
    id_vars=id_cols,
    value_vars=year_cols,
    var_name="year",
    value_name="value"
)

world_long["year"] = world_long["year"].astype(int)
world_long["value"] = pd.to_numeric(world_long["value"], errors="coerce")

display(world_long.head(10))
print(world_long.shape)
```

```python
indicator_map = {
    "NY.GDP.PCAP.CD": "gdp_per_capita",
    "SP.POP.TOTL": "population",
    "SP.DYN.LE00.IN": "life_expectancy",
    "IT.NET.USER.ZS": "internet_users",
    "EG.ELC.ACCS.ZS": "electricity_access",
    "SE.SEC.ENRR": "secondary_school_enrollment",
    "SL.UEM.TOTL.ZS": "unemployment",
    "AG.LND.FRST.ZS": "forest_area"
}

world_long["indicator"] = world_long["Series Code"].map(indicator_map)

display(world_long[["Country Name", "Country Code", "Series Code", "indicator", "year", "value"]].head(10))
```

```python
world_clean = world_long.pivot_table(
    index=["Country Name", "Country Code", "year"],
    columns="indicator",
    values="value"
).reset_index()

world_clean.columns.name = None

world_clean = world_clean.rename(columns={
    "Country Name": "country",
    "Country Code": "country_code"
})

display(world_clean.head(10))
print(world_clean.shape)
```

```python
world_clean = world_clean[
    (world_clean["year"] >= 2000) &
    (world_clean["year"] <= 2024)
]

display(world_clean.head())
print(world_clean["year"].min(), world_clean["year"].max())
```

```python
missing_count = world_clean.isnull().sum()
missing_ratio = world_clean.isnull().mean().sort_values(ascending=False)

display(missing_count)
display(missing_ratio)
```

```python
world_clean.to_csv("world_bank_cleaned.csv", index=False)
```

## 4. Exploratory Data Analysis

### Q1. How are the main development indicators distributed across countries?

```python
indicator_cols = [
    "gdp_per_capita",
    "population",
    "life_expectancy",
    "internet_users",
    "electricity_access",
    "secondary_school_enrollment",
    "unemployment",
    "forest_area"
]

latest_year = world_clean["year"].max()

q1_data = world_clean[world_clean["year"] == latest_year]

q1_long = q1_data.melt(
    id_vars=["country", "country_code", "year"],
    value_vars=indicator_cols,
    var_name="indicator",
    value_name="value"
)

fig = px.box(
    q1_long,
    x="indicator",
    y="value",
    color="indicator",
    facet_col="indicator",
    facet_col_wrap=4,
    points="outliers",
    title=f"Distribution of Development Indicators Across Countries in {latest_year}",
    labels={
        "indicator": " ",
        "value": " "
    },
    color_discrete_sequence=px.colors.qualitative.Set2
)

fig.update_yaxes(matches=None)
fig.update_xaxes(matches=None, showticklabels=False)

fig.update_layout(
    template="plotly_white",
    width=1100,
    height=700,
    title_x=0.5,
    showlegend=False,
    plot_bgcolor="white",
    paper_bgcolor="white"
)

fig.update_traces(
    marker=dict(size=4, opacity=0.65),
    line=dict(width=2)
)

fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))

fig.show()

# I checked how each development indicator is distributed across countries.
# This helps show which indicators have large gaps between countries.
```

### Q2. How have the main development indicators changed over time?


```python
indicator_cols = [
    "gdp_per_capita",
    "population",
    "life_expectancy",
    "internet_users",
    "electricity_access",
    "secondary_school_enrollment",
    "unemployment",
    "forest_area"
]

q2_global = world_clean.groupby("year")[indicator_cols].mean()

full_years = range(world_clean["year"].min(), world_clean["year"].max() + 1)

q2_global = q2_global.reindex(full_years)

q2_global = q2_global.interpolate(
    method="linear",
    limit_direction="both"
)

q2_global = q2_global.reset_index().rename(columns={"index": "year"})

q2_indexed = q2_global.copy()

base_year = q2_indexed["year"].min()

for col in indicator_cols:
    base_value = q2_indexed.loc[q2_indexed["year"] == base_year, col].iloc[0]
    q2_indexed[col] = q2_indexed[col] / base_value * 100

q2_long = q2_indexed.melt(
    id_vars="year",
    value_vars=indicator_cols,
    var_name="indicator",
    value_name="index_value"
)

fig = px.line(
    q2_long,
    x="year",
    y="index_value",
    color="indicator",
    markers=True,
    title=f"Indexed Global Trends of Development Indicators Over Time ({base_year} = 100)",
    labels={
        "year": "Year",
        "index_value": f"Index Value ({base_year} = 100)",
        "indicator": "Indicator"
    },
    color_discrete_sequence=px.colors.qualitative.Set2
)

fig.update_traces(
    line=dict(width=3),
    marker=dict(size=6)
)

fig.update_layout(
    template="plotly_white",
    width=1000,
    height=600,
    title_x=0.5,
    legend_title_text="Indicator",
    plot_bgcolor="white",
    paper_bgcolor="white"
)

fig.add_hline(
    y=100,
    line_width=1,
    line_dash="dash",
    line_color="gray"
)

fig.show()

# I compared how the indicators changed over time.
# I used the first year as 100 to focus on relative change.
```

## 5. Relationships Between Development Dimensions

### Q3. Do countries with better electricity access also have higher life expectancy?

```python
latest_year = world_clean["year"].max()

q3_data = world_clean[
    world_clean["year"] == latest_year
].dropna(subset=["electricity_access", "life_expectancy"])

q3_data["electricity_group"] = pd.cut(
    q3_data["electricity_access"],
    bins=[0, 50, 80, 95, 100],
    labels=["Low", "Medium", "High", "Near universal"],
    include_lowest=True
)

q3_corr = q3_data["electricity_access"].corr(q3_data["life_expectancy"])

fig = px.box(
    q3_data,
    x="electricity_group",
    y="life_expectancy",
    color="electricity_group",
    points="all",
    hover_name="country",
    title=f"Life Expectancy by Electricity Access Level in {latest_year}",
    labels={
        "electricity_group": "Electricity Access Level",
        "life_expectancy": "Life Expectancy (years)"
    },
    category_orders={
        "electricity_group": ["Low", "Medium", "High", "Near universal"]
    },
    color_discrete_sequence=px.colors.qualitative.Set2
)

fig.update_traces(
    marker=dict(size=6, opacity=0.65, line=dict(width=0.5, color="white")),
    line=dict(width=2)
)

fig.update_layout(
    template="plotly_white",
    width=900,
    height=600,
    title_x=0.5,
    showlegend=False,
    plot_bgcolor="white",
    paper_bgcolor="white"
)

fig.add_annotation(
    x=0.02,
    y=0.96,
    xref="paper",
    yref="paper",
    text=f"Correlation: {q3_corr:.2f}",
    showarrow=False,
    font=dict(size=14),
    bgcolor="white",
    bordercolor="gray",
    borderwidth=1
)

fig.show()

# I examined whether countries with better electricity access also have higher life expectancy.
# This shows the relationship between basic infrastructure and health.
```

## 6. Balanced and Unbalanced Development


### Q7. Does environmental development move together with social development over time?

```python
social_cols = [
    "life_expectancy",
    "internet_users",
    "electricity_access",
    "secondary_school_enrollment"
]

q7_global = world_clean.groupby("year", as_index=False)[social_cols + ["forest_area"]].mean()

q7_global[social_cols + ["forest_area"]] = q7_global[social_cols + ["forest_area"]].interpolate(
    method="linear",
    limit_direction="both"
)

q7_normalized = q7_global.copy()

for col in social_cols + ["forest_area"]:
    q7_normalized[col] = (
        (q7_normalized[col] - q7_normalized[col].min()) /
        (q7_normalized[col].max() - q7_normalized[col].min()) * 100
    )

q7_normalized["social_development_score"] = q7_normalized[social_cols].mean(axis=1)

q7_plot = q7_normalized[["year", "social_development_score", "forest_area"]].melt(
    id_vars="year",
    var_name="indicator",
    value_name="normalized_value"
)

q7_plot["indicator"] = q7_plot["indicator"].replace({
    "social_development_score": "Social Development Score",
    "forest_area": "Forest Area"
})

fig = px.line(
    q7_plot,
    x="year",
    y="normalized_value",
    color="indicator",
    markers=True,
    title="Social Development Trend Compared with Forest Area Over Time",
    labels={
        "year": "Year",
        "normalized_value": "Normalized Value",
        "indicator": "Indicator"
    },
    color_discrete_sequence=px.colors.qualitative.Set2
)

fig.update_traces(
    line=dict(width=3),
    marker=dict(size=6)
)

fig.update_layout(
    template="plotly_white",
    width=950,
    height=600,
    title_x=0.5,
    plot_bgcolor="white",
    paper_bgcolor="white"
)

fig.show()

# I compared the social development trend with forest area over time.
# This shows whether environmental conditions move together with social development.
```

### Q8. Which countries show unbalanced development patterns?


```python
development_cols = [
    "gdp_per_capita",
    "life_expectancy",
    "internet_users",
    "electricity_access",
    "secondary_school_enrollment",
    "unemployment",
    "forest_area"
]

year_counts = world_clean.groupby("year")[development_cols].count()

valid_years = year_counts[
    (year_counts[development_cols] > 50).all(axis=1)
].index

q8_year = valid_years.max()

q8_data = world_clean[
    world_clean["year"] == q8_year
].dropna(subset=development_cols).copy()

positive_cols = [
    "gdp_per_capita",
    "life_expectancy",
    "internet_users",
    "electricity_access",
    "secondary_school_enrollment",
    "forest_area"
]

for col in positive_cols:
    q8_data[col + "_score"] = (
        (q8_data[col] - q8_data[col].min()) /
        (q8_data[col].max() - q8_data[col].min()) * 100
    )

q8_data["unemployment_score"] = (
    (q8_data["unemployment"].max() - q8_data["unemployment"]) /
    (q8_data["unemployment"].max() - q8_data["unemployment"].min()) * 100
)

score_cols = [
    "gdp_per_capita_score",
    "life_expectancy_score",
    "internet_users_score",
    "electricity_access_score",
    "secondary_school_enrollment_score",
    "unemployment_score",
    "forest_area_score"
]

q8_data["imbalance_score"] = q8_data[score_cols].std(axis=1)

q8_data["imbalance_group"] = pd.qcut(
    q8_data["imbalance_score"],
    q=4,
    labels=[
        "Low imbalance",
        "Moderate imbalance",
        "High imbalance",
        "Very high imbalance"
    ]
)

color_map = {
    "Low imbalance": "#2ca25f",
    "Moderate imbalance": "#fee08b",
    "High imbalance": "#fdae61",
    "Very high imbalance": "#d7191c"
}

fig = px.choropleth(
    q8_data,
    locations="country_code",
    color="imbalance_group",
    hover_name="country",
    hover_data={
        "imbalance_score": ":.1f",
        "gdp_per_capita_score": ":.1f",
        "life_expectancy_score": ":.1f",
        "internet_users_score": ":.1f",
        "electricity_access_score": ":.1f",
        "secondary_school_enrollment_score": ":.1f",
        "unemployment_score": ":.1f",
        "forest_area_score": ":.1f",
        "country_code": False
    },
    title=f"Countries with Unbalanced Development Patterns in {q8_year}",
    labels={
        "imbalance_group": "Imbalance Level"
    },
    color_discrete_map=color_map,
    category_orders={
        "imbalance_group": [
            "Low imbalance",
            "Moderate imbalance",
            "High imbalance",
            "Very high imbalance"
        ]
    }
)

fig.update_layout(
    template="plotly_white",
    width=950,
    height=550,
    title_x=0.5,
    geo=dict(
        showframe=False,
        showcoastlines=True,
        projection_type="natural earth"
    ),
    plot_bgcolor="white",
    paper_bgcolor="white",
    legend_title_text="Imbalance Level"
)

fig.show()

# I identified countries with large differences among development indicators.
# Darker map colors show countries with more unbalanced development.
```

### Q9. Which countries have the most balanced development profiles?

```python
development_cols = [
    "gdp_per_capita",
    "life_expectancy",
    "internet_users",
    "electricity_access",
    "secondary_school_enrollment",
    "unemployment",
    "forest_area"
]

year_counts = world_clean.groupby("year")[development_cols].count()

valid_years = year_counts[
    (year_counts[development_cols] > 50).all(axis=1)
].index

q9_year = valid_years.max()

q9_data = world_clean[
    world_clean["year"] == q9_year
].dropna(subset=development_cols).copy()

positive_cols = [
    "gdp_per_capita",
    "life_expectancy",
    "internet_users",
    "electricity_access",
    "secondary_school_enrollment",
    "forest_area"
]

for col in positive_cols:
    q9_data[col + "_score"] = (
        (q9_data[col] - q9_data[col].min()) /
        (q9_data[col].max() - q9_data[col].min()) * 100
    )

q9_data["unemployment_score"] = (
    (q9_data["unemployment"].max() - q9_data["unemployment"]) /
    (q9_data["unemployment"].max() - q9_data["unemployment"].min()) * 100
)

score_cols = [
    "gdp_per_capita_score",
    "life_expectancy_score",
    "internet_users_score",
    "electricity_access_score",
    "secondary_school_enrollment_score",
    "unemployment_score",
    "forest_area_score"
]

q9_data["average_development_score"] = q9_data[score_cols].mean(axis=1)
q9_data["development_balance_gap"] = q9_data[score_cols].std(axis=1)
q9_data["balanced_development_score"] = (
    q9_data["average_development_score"] - q9_data["development_balance_gap"]
)

q9_top = q9_data.sort_values(
    "balanced_development_score",
    ascending=False
).head(15)

fig = px.bar(
    q9_top.sort_values("balanced_development_score"),
    x="balanced_development_score",
    y="country",
    orientation="h",
    color="balanced_development_score",
    hover_data={
        "average_development_score": ":.1f",
        "development_balance_gap": ":.1f",
        "balanced_development_score": ":.1f"
    },
    title=f"Countries with the Most Balanced Development Profiles in {q9_year}",
    labels={
        "balanced_development_score": "Balanced Development Score",
        "country": "Country"
    },
    color_continuous_scale="Blues"
)

fig.update_layout(
    template="plotly_white",
    width=950,
    height=650,
    title_x=0.5,
    plot_bgcolor="white",
    paper_bgcolor="white",
    coloraxis_showscale=False
)

fig.update_traces(
    marker_line_width=0.5,
    marker_line_color="white"
)

fig.show()

# I identified countries with the most balanced development profiles.
# These countries perform consistently well across multiple development dimensions.
```

## 7. Overall Relationship Summary

### Q10. What relationships exist among all development indicators?

```python
indicator_cols = [
    "gdp_per_capita",
    "population",
    "life_expectancy",
    "internet_users",
    "electricity_access",
    "secondary_school_enrollment",
    "unemployment",
    "forest_area"
]

year_counts = world_clean.groupby("year")[indicator_cols].count()

valid_years = year_counts[
    (year_counts[indicator_cols] > 50).all(axis=1)
].index

q10_year = valid_years.max()

q10_data = world_clean[
    world_clean["year"] == q10_year
][indicator_cols].dropna()

q10_corr = q10_data.corr()

fig = px.imshow(
    q10_corr,
    text_auto=".2f",
    color_continuous_scale="RdBu",
    zmin=-1,
    zmax=1,
    title=f"Correlation Matrix of Development Indicators in {q10_year}",
    labels={
        "color": "Correlation"
    }
)

fig.update_layout(
    template="plotly_white",
    width=850,
    height=800,
    title_x=0.5,
    plot_bgcolor="white",
    paper_bgcolor="white"
)

fig.show()

# I summarized the relationships among all indicators using a correlation heatmap.
# This shows which development indicators move together and which are more independent.
```

## 8. Conclusion

## The main finding is that development indicators are connected, but not perfectly aligned. GDP is important, but it does not fully represent health, education, technology access, employment, infrastructure, and environmental conditions. Therefore, national development should be evaluated as a multidimensional and balanced concept.
